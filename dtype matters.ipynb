```python
import glob
from scipy.ndimage import imread
from scipy.misc import imresize
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

读取图片，然后转化为ndarray
查看数据类型： **dtype('uint8')**

```python
im = Image.open('00000133.jpg')
img = np.array(im)
img.dtype
```

```python
def normalize(img, cdim=3):
#     [0-255]/255 * 2-1 ---->[-1,1]
    img.astype(float)
    return img/255.0*2-1
```

对一个4-d数组做归一化（num, height,width,channel)
按照 `b=np.ones_like(a)`定义的b,其数据类型是 uint8,

所有最后得到的b是0

```python
a = np.array([img,img])
b = np.ones_like(a)
for i,ai in enumerate(a):
    b[i] = normalize(ai)
```

```python
b.dtype
```

```python
b[0]
```

```python
plt.imshow(b[0])
```

要的得到想要的b，需要如下定义 `b = np.ones_like(a,dtype=np.float)`

```python
b = np.ones_like(a,dtype=np.float)
for i,ai in enumerate(a):
    b[i] = normalize(ai)
```
