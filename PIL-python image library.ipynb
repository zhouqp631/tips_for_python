```python
%matplotlib inline
import os, sys
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image,ImageFilter
```

```python
ls *.jpeg
```

```python
infile = './space.jpeg'
im = Image.open(infile)
# 查看图片的格式 大小 和 模式(L,RGB, CMKY)
im.format,im.size,im.mode
```

```python
plt.imshow(im)
```

## 图片保存为其他格式

```python
f,e = os.path.splitext(infile)
im.save(f+'.png')
%ls dog*
```

## 分离颜色通道

```python
r,g,b = im.split()
rr = np.array(r)
plt.imshow(rr)
```

```python
bb = np.array(b)
plt.imshow(bb)
```

## 一些滤波-filter

```python
plt.imshow(im.filter(ImageFilter.BLUR))
```

```python
# 检测边缘
plt.imshow(im.filter(ImageFilter.FIND_EDGES))
```
